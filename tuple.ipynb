```python
single_element_tuple = (42,)
single_element_string_tuple = ("hello",)
multiple_elements_tuple = (1,"apple", 3.14, True)

print("Single element integer tuple:", single_element_tuple)
print("Single element string tuple:", single_element_string_tuple)
print("Multiple elements tuple:", multiple_elements_tuple)
```

```python
print(multiple_elements_tuple[0])

print(multiple_elements_tuple[2])
```

```python
print(multiple_elements_tuple[1:3])

print(multiple_elements_tuple[:2])

print(multiple_elements_tuple[2:])

print(multiple_elements_tuple[::2])

print(multiple_elements_tuple[::-1])
```

```python
# Demonstrate positive indexing
print("Accessing the first element using positive indexing:", multiple_elements_tuple[0])
print("Accessing the third element using positive indexing:", multiple_elements_tuple[2])
```

```python
# Demonstrate negative indexing
print("Accessing the last element using negative indexing:", multiple_elements_tuple[-1])
print("Accessing the second to last element using negative indexing:", multiple_elements_tuple[-2])
```

```python
print("Accessing element from index 1 to index 3(exclusive) using slicing:",multiple_elements_tuple[1:3])
print("Accessing elements from the beginning to index 2 (exclusive) using slicing:", multiple_elements_tuple[:2])
print("Accessing elements from index 2 to the end using slicing:", multiple_elements_tuple[2:])
print("Accessing every second element using slicing:", multiple_elements_tuple[::2])
print("Accessing elements in reverse order using slicing:", multiple_elements_tuple[::-1])
```

```python

# Create a tuple representing a point in 3D space (x, y, z)
point = (10, 25, 5)

# Access elements using positive indexing
x_coord = point[0]
y_coord = point[1]

# Access the last element using negative indexing
z_coord = point[-1]

# Access a slice of the tuple
xy_coords = point[0:3]

print(f"Point tuple: {point}")
print(f"X-coordinate: {x_coord}")
print(f"Y-coordinate: {y_coord}")
print(f"Z-coordinate: {z_coord}")
print(f"XY-coordinates (slice): {xy_coords}")
```

```python
# Create a tuple of temperatures over a week
weekly_temperatures = (75, 80, 78, 82, 80, 79, 80)

# Get the number of days
num_days = len(weekly_temperatures)

# Count how many times 80 degrees occurred
count_80 = weekly_temperatures.count(80)

# Find the index of the first day with 78 degrees
index_78 = weekly_temperatures.index(78)

print(f"Weekly Temperatures: {weekly_temperatures}")
print(f"Number of days: {num_days}")
print(f"Count of 80 degrees: {count_80}")
print(f"Index of the first 78 degrees: {index_78}")
```

```python
# Example 1: Creating and Accessing a Tuple

# Create a tuple representing a point in 3D space (x, y, z)
point = (10, 25, 5)
print(f"Point tuple: {point}")

# Access elements using positive indexing
x_coord = point[0]
y_coord = point[1]
print(f"X-coordinate (point[0]): {x_coord}")
print(f"Y-coordinate (point[1]): {y_coord}")

# Access the last element using negative indexing
z_coord = point[-1]
print(f"Z-coordinate (point[-1]): {z_coord}")

# Access a slice of the tuple
xy_coords = point[0:2]
print(f"XY-coordinates (slice point[0:2]): {xy_coords}")

print("-" * 30) # Separator

# Example 2: Using Built-in Functions and Methods

# Create a tuple of temperatures over a week
weekly_temperatures = (75, 80, 78, 82, 80, 79, 80)
print(f"Weekly Temperatures: {weekly_temperatures}")

# Get the number of days using len()
num_days = len(weekly_temperatures)
print(f"Number of days (len(weekly_temperatures)): {num_days}")

# Count how many times 80 degrees occurred using count()
count_80 = weekly_temperatures.count(80)
print(f"Count of 80 degrees (weekly_temperatures.count(80)): {count_80}")
```
