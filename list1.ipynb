```python
my_list =[1,2,3]
new_list =[1,2,3,4,5]
my_list.append(4)

print(f"after append(4):{my_list}")
my_list.append(new_list)
print(f"after append(new_list):{my_list}")

my_list.insert(2,4.4)
print(f"after append(4.4):{my_list}")
new_list =[1,2,3,4,5]

my_list.extend(new_list)
print(f"after extend([1,2,3,4,5]):{my_list}")
```

```python
# nested list
nested_list=[1,2,['a','b','c'],4,[True,False]]
print(f"original nested list: {nested_list}")
print(f"Element at index 0:{nested_list[0]}")
print(f"Element at index 2:{nested_list[2]}")

print(f"Element at index 1 of the list at index 2: {nested_list[2][1]}")
print(f"Element at index 0 of the list at index 4: {nested_list[4][0]}")

nested_list[2][2]='d'
print(f"After modifying element:{nested_list}")

nested_list[4].append(True)
print(f"After appending to nested list:{nested_list}")
nested_list[4].remove(True)
print(f"After remove {nested_list}")
```
