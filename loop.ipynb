27 WAP to print number from 100 to 1. Print 10 numbers on each line

```python
i=100
while i>0:
  print(i,end=" ")
  i-=1
  if i%10==0:
   print("\n")
```

28 WAP to print all odd integers from 13 to 69. Print 8 integers on each line

```python
i=13
count=0
while i<70:
  if i%2!=0:
    print(i,end=" ")
  i+=1
  count+=1
  if count%16==0:
    print("\n")
```
