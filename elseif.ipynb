A gaming cafe charges different rates based on total hours played. From 0 to 2 hours the rate is 120
per hour. From more than 2 to 5 hours the rate is 100 per hour. More than 5 hours the rate is 80 per
hour. Given the number of hours played determine the total bill amount.

```python
cafe=int(input("Enter played houres"))
if cafe<=2:
  amount= cafe*120
  print(f"your payment is: {amount}")
elif cafe>2 and cafe <=5:
  amount= cafe*100
  print(f"your payment is:{amount}")
elif cafe>5:
  amount= cafe*80
  print(f"your payment is: {amount}")
else:
  print("Thank you")
```

A student grade depends on marks and attendance. If marks are 90 or above and attendance is 95 or
above the grade is A. If marks are 80 or above or attendance is 90 or above the grade is B. If marks are
60 or above the grade is C. Otherwise the grade is D. Given marks and attendance determine the
grade.

```python
marks=float(input("Enter your marks"))
att=float(input("Enter your attendence"))
if marks>=90 and att>=95:
  print("A grade")
elif marks>=80 and att>=90:
  print("B grade")
elif marks>=60 and att>=70:
  print("C grade")
else:
    print("D grade")
```

A shop offers discounts based on the purchase amount. Less than 500 gives no discount. 500 to 999
gives 5 percent discount. 1000 to 4999 gives 10 percent discount. 5000 or more gives 20 percent
discount. Given the amount calculate the final payable amount.

```python
pamount=float(input("Enter purchase amount"))
if pamount<500:
  print("no discount")
elif pamount>=500 and pamount<=999:
  dis=(5/100)*pamount
  amo=pamount - dis
  print(f"5 percent discount applied your total bill is {amo}")
```
