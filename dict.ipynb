```python
empty_dict_braces ={}
print(f"using curly braces:{empty_dict_braces}")
empty_dict_constructor =dict()
print(f"using dict() constructor:{empty_dict_constructor}\n")
print("2.creating a dictionary with initial key-value pairs using curly braces:")
person_data ={
    'name':'Akash',
    'age' :30,
    'city':'delhi'
}
print(f"person data: {person_data}\n")

print("3.Creating a dictionary using the dict() constructor with keyword arguments:")
car_data = dict(make='Bmw', model='m5',year=2022)
print(f"car data: {car_data}\n")

print("4. Creating a dictionary from a list of key-value tuples:")
fruit_prices_list = [('apple',10.5),('banana',20.3),('pineapple',11)]
print(f"fruit_prices_list:{fruit_prices_list}\n")
```

```python
# ("---Accessing Dictionary Elements---")

student_info={
    'name':'Akash',
    'age': 22,
    'major':'data analyist'
}
print(f"original dictionary: {student_info}\n")

# 1. Demonstrate accessing a dictionary value using its key
#("1. Accessing existing values using direct key access:")
student_name = student_info['name']
print(f"student name:{student_name}")
print(student_info['age'])
print("-"*50)

# 2. Introduce the get() method as a safer way to access values.
# Show how it returns None by default if the key is not found.
#(". Accessing values using the .get() method (safer):")
student_major = student_info.get('major')
student_gpa = student_info.get('gpa') # 'gpa' does not exist
print(f"student_major:{student_major}")
print(f"Student's GPA (using .get() for non-existent key: {student_gpa}\n")

# 3. Show how to provide a custom default value to the get() method
# when the key is not found.

#("4. Using .get() with a custom default value:")
student_city = student_info.get('city','not specified')
student_age_with_default = student_info.get('age',0)
print(f"student_city:{student_city}")
print(f"student age(with default with not found):{student_age_with_default}\n")
```
