```python
my_set = set()
print(f"empty set:{my_set}")

my_set = {1,2,3,2,4,5}
print(f"set with elements:{my_set}")
list_data =[10,20,30,20]
set_from_list= set(list_data)
print(f"set from list: {set_from_list}")
string_data = "hello"
set_from_list = set(string_data)
print(f"set from list: {set_from_list}")
```

```python
my_set ={1,2,3}
my_set.add(4)
print(f"after add 4:{my_set}")
my_set.add(2)
print(f"after adding 2 again:{my_set}")
```

```python
my_set ={1,2,3,4}
my_set.remove(3)
print(f"after removing 3: {my_set}")
```

```python
my_set ={1,2,3,4}
my_set.discard(2)
print(f"after discard 2:{my_set}")
my_set.discard(5)
print(f"after discard 5:{my_set}")
```

```python
my_set ={1,2,6,4,5}
print(my_set)
pop_element=my_set.pop()
print(f"pop element:{pop_element}")
print(f"set after pop:{my_set}")
```
